## 4. Assess quality of the data using standard data quality dimensions

### 4.2 Uniqueness

```python
print("Number of duplicated values in the data = ", df.duplicated().sum())
```

### 4.3 consistency

```python
for column in df.columns:
    if df[column].dtype == "object":
        print(f"Value Counts of", df[[column]].value_counts())
        print("\n")
```

### 4.4 Currency

```python
print("The minimum date in tpep_pickup_datetime column:", df.tpep_pickup_datetime.min())
print("The maximum date in tpep_pickup_datetime column:", df.tpep_pickup_datetime.max())
```

```python
print("The minimum date in tpep_dropoff_datetime column:", df.tpep_dropoff_datetime.min())
print("The maximum date in tpep_dropoff_datetime column:", df.tpep_dropoff_datetime.max())
```

### 4.6 Validity

```python
for column in df.columns:
    if df[column].dtype == 'int':
        print(f"Value Counts of", df[[column]].value_counts(sort=False))
        print("\n")
```

```python
for column in df.columns:
    if df[column].dtype == 'float':
        print(f"Value Counts of", df[[column]].value_counts(sort=False))
        print("\n")
```
